```python
from word2number import w2n
```

```python
df.columns
```

```python
df = df.rename(columns={'test_score(out of 10)' : 'test_score', 'interview_score(out of 10)' : 'interview_score','salary($)' : 'salary'})
```

```python
df.columns
```

```python
df = df.fillna(df.mode())
```

```python
df['experience_num'] = df['experience'].apply(w2n.word_to_num)
df.head()
```

```python
df_final = df.drop(columns=['experience'], axis=1)
```

```python
df_final.columns
```

```python
X = df_final.drop(columns='salary', axis = 0)
y = df_final['salary']
```

```python
X = X.fillna(X.median())
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.predict([[8,8,11]])
```

```python
lr.predict([[9,8,5]])
```

```python
data = {
    'test_score':[8,9,5,8],
    'interview_score':[6,80,7,6],
    'experience_num':[9,5,2,1]
}
predict_df = pd.DataFrame(data)
predict_df.head()
```

```python
predict_df['salary'] = pd.DataFrame(lr.predict(predict_df))
```
